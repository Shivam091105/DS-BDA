```python
age_list_by_gender = df.groupby("gender")["age"].apply(list)

age_list_by_gender
```

```python
grouped_stats = df.groupby("gender")["age"].agg(
    Mean="mean",
    Median="median",
    Minimum="min",
    Maximum="max",
    Standard_Deviation="std"
)

grouped_stats
```

```python
df["income_numeric"] = df["income"].map({"<=50K": 0, ">50K": 1})
```

```python
grouped_income = df.groupby(["education", "occupation"])['income_numeric'].agg(
    Mean = "mean", Median="median", Minimum="min", Maximum="max", Standard_Deviation="std"
)
grouped_income
```

```python
grouped_gain = df.groupby("gender")['capital-gain'].agg(
    Total_Gain = "sum",
    Avg_Gain = "mean",
    Count = "count"
)
grouped_gain
```

```python
grouped_income = df.groupby(["occupation", "education"])['income_numeric'].agg(
    Mean = "mean", Median="median", Minimum="min", Maximum="max", Standard_Deviation="std"
)
grouped_income
```
