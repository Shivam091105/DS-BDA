```python

  import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Global theme settings
sns.set_theme(
    style="whitegrid",
    context="talk",
    palette="Set1"
)
```

```python
sns.countplot(x="sex", hue="survived", data=df)
plt.title("Survival Count by Gender")
plt.show()
```

```python
sns.countplot(x="pclass", hue="survived", data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
sns.boxplot(x="survived", y="fare", data=df)
plt.title("Fare vs Survival")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['fare'], bins=30, kde=True)

plt.title("Distribution of Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
g = sns.FacetGrid(
    df,
    col="survived",
    row="sex",
    height=4,
    aspect=1
)

g.map(sns.histplot, "age", bins=30, kde=True)
g.set_axis_labels("Age", "Count")
g.set_titles(row_template="{row_name}", col_template="Survived: {col_name}")

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.boxplot(
    x="sex",
    y="age",
    hue="survived",
    data=df,
    showfliers=False
)

sns.stripplot(
    x="sex",
    y="age",
    hue="survived",
    data=df,
    dodge=True,
    alpha=0.3,
    size=3,
    jitter=True
)

plt.title("Age Distribution with Individual Data Points")
plt.xlabel("Gender")
plt.ylabel("Age")

ax = plt.gca()
ax.legend_.remove()  # avoid duplicate legends
plt.show()
```
