```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
sns.set_theme(
    style="whitegrid",
    context="talk",
    palette="Set1"
)
plt.rcParams["figure.figsize"] = (8,5)
```

```python
sns.boxplot(x="species", y="sepal_length",hue="species", data=df)
plt.title("Sepal Lengths as per Species")
plt.show()
```

```python
sns.boxplot(x="species", y="sepal_width",hue="species", data=df)
plt.title("Sepal Widths as per Species")
plt.show()
```

```python
sns.boxplot(x="species", y="petal_length",hue="species", data=df)
plt.title("Petal Lengths as per Species")
plt.show()
```

```python
numeric_features = df.drop(columns='species').columns

for feature in numeric_features:
    plt.figure()
    plt.hist(df[feature])
    plt.title(f"Histogram of {feature}")
    plt.show()

for feature in numeric_features:
    plt.figure()
    plt.boxplot(df[feature])
    plt.title(f"Boxplot of {feature}")
    plt.show()
```

```python
sns.boxplot(x="species", y="petal_width",hue="species", data=df)
plt.title("Sepal Lengths as per Species")
plt.show()
```

```python
for column in df.columns[:-1]:
    plt.figure()
    sns.histplot(df[column], kde=True, bins=20)
    plt.title(f"Distribution of {column}")
    plt.show()
```

```python
sns.swarmplot(x="species", y="petal_length", data=df)
plt.title("Swarm Plot of Petal Length")
plt.show()
```

```python
sns.pairplot(df, hue="species")
plt.show()
```
